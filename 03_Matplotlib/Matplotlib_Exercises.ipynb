Conduct a Univariate Analysis of the participants' Cholesterol. What is noticeable about it? Create a function that normalizes the participants' Cholesterol and create two visualizations, one before the `"Chol"` data is modified, and one after normalizing the `"Chol"` data.

```python
q1 = data['Chol'].quantile(0.25)
q3 = data['Chol'].quantile(0.75)
iqr = q3 - q1
data_chol_norm = data[(data.Chol > (q1 - iqr)) & (data.Chol < (q3 + iqr))].Chol
print(data['Chol'].size)
print(data_chol_norm.size)
```

```python
# YOUR CODE HERE
plt.figure(1)
plt.subplot(1,2,1)
sns.distplot(data['Chol'], kde=True)
plt.xlabel('Cholesterol')
plt.ylabel('% of people')
plt.title('Original Cholesterol Data')

plt.subplot(1,2,2)
data['Chol norm'] = (data_chol_norm - data_chol_norm.mean()) / data_chol_norm.std()
sns.distplot(data['Chol norm'], kde=True)
plt.xlabel('Cholesterol')
plt.ylabel('% of people')
plt.title('Normalized Cholesterol Data')
```

```python
data["Chol_norm"] = data_chol_norm/data_chol_norm.abs().max()
data.Chol_norm.hist()
```

Fill in the function below. The function recieves a non-empty string and an int n, and returns a new string where the character at index n has been removed.¶


```python
def remove_char(old_string, n):
    return old_string[0:n] + old_string[n+1:]
```

```python
print(remove_char("h", 0))
print(remove_char("hi", 0))
print(remove_char("hi", 1))
print(remove_char("hello", 3))
print(remove_char("hello", -4))
```

Given a string, return a new string where the first and last characters have been exchanged.

```python
def swap_chars(old_string):
  return old_string[-1:] + old_string[1:-1] + old_string[:1]
```

Test the function with three different parameters


```python
# YOUR CODE HERE
print(swap_chars("test"))
print(swap_chars("hello"))
print(swap_chars("medlytics"))
```
